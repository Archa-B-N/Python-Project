# Preprocessing Steps
The following steps were taken to preprocess the data:

1. Import necessary libraries: pandas, numpy, matplotlib, and seaborn were imported to facilitate data manipulation and visualization.

2. Load the dataset: The Excel file was loaded using pd.read_excel().

3. Check for missing values: df.isnull().sum() was used to identify missing values in the dataset.

4. Handle missing values: A strategy was decided upon to handle missing values, such as mean/median imputation or removal.

5. Correct Height column: The Height column was corrected by replacing height values with numbers between 150-180 using df['Height'].apply().

```python
#Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Load dataset
df = pd.read_excel('myexcel.xlsx')

#Check for missing values
print("Missing Values:")
print(df.isnull().sum())

#Correct Height column: replace with random values between 150-180
df['Height'] = df['Height'].apply(lambda x: np.random.randint(150, 180))

#Verify corrections
print("\nHeight Column Corrections:")
print(df['Height'].describe())

```

# Analysis Tasks:
# 1. Determine the distribution of employees across each team and calculate the percentage split relative to the total number of employees. 


```python
#Step 2: Calculate team distribution


# Calculate team distribution
team_distribution = df['Team'].value_counts()

# Calculate percentage split
percentage_split = (team_distribution / len(df)) * 100

# Combine results
team_distribution_percentage = pd.concat([team_distribution, percentage_split], axis=1)

# Rename columns
team_distribution_percentage.columns = ['Count', 'Percentage']

# Print results
print(team_distribution_percentage)
```

# NBA Players Team Distribution Chart

```python
# Visualize team distribution 

# Bar chart
plt.figure(figsize=(20, 9))
plt.bar(team_distribution_percentage.index, team_distribution_percentage['Count'],color='#87CEEB')
plt.xlabel('Team')
plt.ylabel('Count')
plt.title('Team Distribution')
plt.xticks(rotation=80)
plt.show()

```

# Data story based on the team distribution and percentage split:


The NBA comprises 30 teams, each with a unique roster. 

Our analysis reveals that the New Orleans Pelicans have the highest number of players, with 19 players on their roster, accounting for 4.15% of the total NBA players.

The Memphis Grizzlies follow closely, with 18 players, making up 3.93% of the total NBA players.

Interestingly, 24 teams have 15 players each, indicating a relatively even distribution of players across the league.

The Orlando Magic and Minnesota Timberwolves have the lowest number of players, with 14 players each, accounting for 3.06% of the total NBA players.

These insights provide a fascinating glimpse into the world of NBA team dynamics and player distribution.

# 3. Identify the predominant age group among employees. 

```python
# Calculate age distribution


age_distribution = df['Age'].value_counts(bins=[20, 30, 40, 50, 60], sort=False)
'''The code age_distribution = df['Age'].value_counts(bins=[20, 30, 40, 50, 60], sort=False) means:

Breaking down the code:

- df['Age']: Selects the 'Age' column from the DataFrame df.
- value_counts(): Counts the number of occurrences of each unique value in the 'Age' column.
- bins=[20, 30, 40, 50, 60]: Instead of counting individual ages, group them into bins (or ranges) defined by these values. The bins are:
    - 20-29
    - 30-39
    - 40-49
    - 50-59
- sort=False: By default, value_counts() sorts the results in descending order. Setting sort=False preserves the original order of the bins.'''

# Calculate percentage split
percentage_split = (age_distribution / len(df)) * 100

# Combine results
age_distribution_percentage = pd.concat([age_distribution, percentage_split], axis=1)

# Rename columns
age_distribution_percentage.columns = ['Count', 'Percentage']

# Print results
print(age_distribution_percentage)

```

# 4. Discover which team and position have the highest salary expenditure. 

```python
team_salary_expenditure = df.groupby('Team')['Salary'].sum().reset_index()
team_salary_expenditure = team_salary_expenditure.sort_values(by='Salary', ascending=False)
team_salary_expenditure
```

#### Insights:

The Cleveland Cavaliers have the highest salary expenditure at approximately $111.8 million.

The top 5 teams with the highest salary expenditure are:

1. Cleveland Cavaliers ($111.8M)

2. Memphis Grizzlies ($95.9M)

3. Los Angeles Clippers ($94.9M)

4. Oklahoma City Thunder ($93.8M)

5. Miami Heat ($92.2M)

This data provides valuable insights into the financial commitments of each team.

```python
# Position salary expenditure
position_salary_expenditure = df.groupby('Position')['Salary'].sum().reset_index()
position_salary_expenditure = position_salary_expenditure.sort_values(by='Salary', ascending=False)
position_salary_expenditure
```

#### Insights:
The Center (C) position has the highest salary expenditure at approximately $466.4 million.

The top 5 positions with the highest salary expenditure are:

1. Center (C) ($466.4M)

2. Point Guard (PG) ($466.2M)

3. Power Forward (PF) ($457.1M)

4. Small Forward (SF) ($412.9M)

5. Shooting Guard (SG) ($411.5M)

```python
# Pivot table
salary_pivot = df.pivot_table(index='Team', columns='Position', values='Salary', aggfunc='mean')
salary_pivot
```

# Visualisation of team and position have the highest salary expenditure

```python

# Graphical representation of Team salary expenditure 
plt.figure(figsize=(10, 6))
plt.bar(team_salary_expenditure['Team'], team_salary_expenditure['Salary'])
plt.xlabel('Team')
plt.ylabel('Salary Expenditure')
plt.title('Team Salary Expenditure')
plt.xticks(rotation=80)
plt.show()

# Graphical representation of Position salary expenditure
plt.figure(figsize=(14, 9))
plt.bar(position_salary_expenditure['Position'], position_salary_expenditure['Salary'])
plt.xlabel('Position')
plt.ylabel('Salary Expenditure')
plt.title('Position Salary Expenditure')
plt.show()

# Heat map (A heat map is a graphical representation of data where values are depicted by color)
plt.figure(figsize=(10, 8))
sns.heatmap(salary_pivot, annot=True, cmap='Blues')
plt.xlabel('Position')
plt.ylabel('Team')
plt.title('Salary Distribution')
plt.show()
```

# 5.Investigate if there's any correlation between age and salary, and represent it visually. 

```python
# Calculate the correlation between Age and Salary
correlation_coefficient = df['Age'].corr(df['Salary'])
print("Correlation Coefficient:", correlation_coefficient)
```

# Graphical representation of correlation between age and salary

```python
# Create a scatter plot of Age vs Salary
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Salary', data=df)
plt.title('Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

# Data story based on correlation between Age and Salary

### Based on the scatter plot, here are some insights:

1. Weak positive correlation: the correlation coefficient is 0.21, indicating a weak positive relationship between age and salary.

2. Age distribution: The age distribution appears to be relatively normal, with most players between 20-30 years old.

3. Outliers: There may be some outliers, such as older players with lower salaries or younger players with higher salaries.

4. No clear trend: The plot doesn't show a clear trend or pattern, which is consistent with the weak correlation coefficient.

These insights can help you better understand the relationship between age and salary in the NBA.
